# NNLS and MCR-ALS fits with water removal

This notebook builds on `FirstFits.ipynb` by adding a water absorption spectrum as a removal term.

We start with some "trusted" spectra at 6 discrete wavelengths from tubes 1-3 that are assumed to be entirely due to emission of particles 1-3, and a negated water spectrum. We then postulate that other regions of interest (ROIs) can be approximated as a weighted sum of those spectra. This amounts to solving

$$ Ax=b $$

where $A$ is the 6x4 (spectra X species) array of spectral shapes, $x$ are the (4x1) unknown weights in each ROI, and $b$ is the provided spectrum (6x1) in the given ROI. We use `nnls` from `scipy.optimize` to do the least squares inversion.

Instructions for getting the below libraries setup are in the `README.md` at the root level. Make sure to set the kernel to `QDimages` (or whatever environment name you installed spectrochempy into) at the top right.

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from scipy.optimize import nnls
import spectrochempy as scp
```

```python
df_ref_w = pd.DataFrame({ 'Tube1': [1423.39, 1141.72, 228.56, 88.83, 2.69, 5.32],
                        'Tube2': [106.93, 355.45, 1436.32, 763.09, 31.24, 34.49],
                        'Tube3': [338.49, 65.13, 119.99, 344.86, 1149.40, 1393.15],
                        'WaterAbs': [-0.2, -1.0, -1.35, -3.9, -20, -10]})

df_ref_3 = pd.DataFrame({ 'Tube1': [1423.39, 1141.72, 228.56, 88.83, 2.69, 5.32],
                        'Tube2': [106.93, 355.45, 1436.32, 763.09, 31.24, 34.49],
                        'Tube3': [338.49, 65.13, 119.99, 344.86, 1149.40, 1393.15]})

A = df_ref_w.to_numpy()
df_ref_w.plot(title="Reference Spectra")
```

```python
def make_image(ax_in, concentrations, reshape):
    ps = np.reshape(concentrations, reshape)
    ax_in.imshow(ps, cmap='gray')
```

```python
tube_array = np.load('../data/tube_input_matrix.npy')

# Apply a function across each row of the numpy array
tube_conc = np.apply_along_axis(lambda b: nnls(A,b)[0], 1, tube_array)
residuals = np.apply_along_axis(lambda b: nnls(A,b)[1], 1, tube_array)

# Convert the weights to an WxL image
tube_shape = (108,344)
tfig, ax = plt.subplots(5,1, figsize=(12,14))
make_image(ax[0], tube_conc[:,0], tube_shape) # Species 1
make_image(ax[1], tube_conc[:,1], tube_shape) # Species 2
make_image(ax[2], tube_conc[:,2], tube_shape) # Species 3
make_image(ax[3], tube_conc[:,3], tube_shape) # Water
make_image(ax[4], residuals, tube_shape) # Residuals
```

```python
mouse_array = np.load('../data/mouse_input_matrix.npy')

# Apply a function across each row of the numpy array
mouse_conc = np.apply_along_axis(lambda b: nnls(A,b)[0], 1, mouse_array)
residuals = np.apply_along_axis(lambda b: nnls(A,b)[1], 1, mouse_array)

# Convert the weights to an WxL image
mouse_shape = (504,368)
# Convert the weights to an WxL image
mfig, ax = plt.subplots(2,3, figsize=(14,10))
make_image(ax[0][0], mouse_conc[:,0], mouse_shape)
make_image(ax[0][1], mouse_conc[:,1], mouse_shape)
make_image(ax[0][2], mouse_conc[:,2], mouse_shape)
make_image(ax[1][0], mouse_conc[:,3], mouse_shape)
make_image(ax[1][1], residuals, mouse_shape)

# Make a plot showing how the water coefficients are distributed
numbins = 100
logbins = np.logspace(0,np.log(np.max(mouse_conc[:,3])),numbins,base=np.e)
(nums, bds) = np.histogram(mouse_conc[:,3], bins=logbins)
ax[1][2].hist(mouse_conc[:,3], bins=logbins)
ax[1][2].set_xscale('log')
```

```python
# Find out which bin each pixel belongs to
depth_bin = np.digitize(mouse_conc[:,3], logbins)
ps = np.reshape(mouse_conc[:,2], mouse_shape)
dbs = np.reshape(depth_bin, mouse_shape)
mouse_shape_3d = (504,368, numbins+1)
c3d = np.zeros(mouse_shape_3d)

for i in range(mouse_shape_3d[0]):
    for j in range(mouse_shape_3d[1]):
        c3d[i,j,dbs[i,j]] = ps[i,j]
```

## Refitting with spectrochempy

We learned that there are some important options to send to the solver when trying to add *and* subtract spectra. However, after changing a bunch of the options, none of the spectra ended up being believable. 

```python
def fit_mcrals(np_file, St_guess):
    X = np.load(np_file)
    #mcr = scp.MCRALS(log_level="INFO", nonnegSpec=[0,1,2], normSpec=None, solverConc='nnls', closureConc=[0,1,2],unimodConc=[])
    mcr = scp.MCRALS(log_level="INFO", nonnegSpec=[0,1,2], solverConc='nnls',solverSpec='pnnls', unimodConc=[],unimodSpec=[0,1])
    mcr.fit(X,St_guess)
    return mcr
```

```python
mcr_mouse = fit_mcrals('../data/mouse_input_matrix.npy', A.T)
mcr_mouse.parameters()
```

```python
mouse_conc = np.array(mcr_mouse.C)
mouse_shape = (504,368)
# Convert the weights to an WxL image
mfig, ax = plt.subplots(2,3, figsize=(14,10))
make_image(ax[0][0], mouse_conc[:,0], mouse_shape)
make_image(ax[0][1], mouse_conc[:,1], mouse_shape)
make_image(ax[0][2], mouse_conc[:,2], mouse_shape)
make_image(ax[1][0], mouse_conc[:,3], mouse_shape)
```
