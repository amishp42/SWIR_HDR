Quick test to see if a simple ratio of the radiance maps with different excitation or emission wavelengths from the same mouse/imaging session shows a difference in depths. It does not. Will require actually doing the experiments right! smile

```python
import numpy as np
import matplotlib.pyplot as plt

# Load the two .npy arrays
array1 = np.load('./data/240515_2hr p.i. RO right side re/2hr p.i. RO right side re_670_BP1350__radiance_map.npy')
array2 = np.load('./data/240515_2hr p.i. RO right side re/2hr p.i. RO right side re_808_BP1350__radiance_map.npy')

# Calculate percentage change
percent_change = np.where(array1 != 0, (array2 - array1) / array1 * 100, 0)

# Create a figure with 3 subplots
fig, (ax1, ax2, ax3) = plt.subplots(1, 3, figsize=(15, 5))

# Display the first original image
im1 = ax1.imshow(array1)
ax1.set_title('Original Image 1')
plt.colorbar(im1, ax=ax1)

# Display the second original image
im2 = ax2.imshow(array2)
ax2.set_title('Original Image 2')
plt.colorbar(im2, ax=ax2)

# Display the percentage change image
im3 = ax3.imshow(percent_change, cmap='RdBu_r', vmin=-10, vmax=10)
ax3.set_title('Percentage Change')
plt.colorbar(im3, ax=ax3, label='% Change')

# Adjust layout and display the plot
plt.tight_layout()
plt.show()
```
