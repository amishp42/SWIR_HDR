```python
import numpy as np
import pandas as pd
from PIL import Image

```

```python
path = "data\\2hr p.i. RO right side re_808_BP1200_denoised.npy"
path2 = "data\\2hr p.i. RO right side re_808_BP1200__radiance_map.npy"

image = np.load(path)
image2 = np.load(path2)

for i in range(image.shape[0]):
    im = Image.fromarray(np.squeeze(image[i,:,:]))
    im.save("data\\2hr p.i. RO right side re_808_BP1200_denoised_{}.tif".format(i), "TIFF")

im2 = Image.fromarray(image2, mode='F')
im2.save("data\\2hr p.i. RO right side re_808_BP1200_radiance_map.tif", "TIFF")


```
